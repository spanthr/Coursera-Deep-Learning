```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```
# GRADED FUNCTION: house_model
def house_model(y_new):
    xs = np.array([0.0,01.0,02.0,05.0,08.0,11.0,09.0], dtype=float)# Your Code Here#
    ys = np.array([.50,1.00,1.50,3.50,4.50,6.00,5.00], dtype=float) # Your Code Here#
    model = model = tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])]) 
            
    model.compile(optimizer='sgd', loss='mean_squared_error')
    model.fit(xs, ys, epochs=500) 
    return model.predict(y_new)[0]
```

```
prediction = house_model([7.0])

print(prediction)
```
